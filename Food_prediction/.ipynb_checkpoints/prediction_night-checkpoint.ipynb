```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import re
```

```python
train_data=pd.read_csv('./data/train.csv')
test_data=pd.read_csv('./data/test.csv')
```

```python
train_data.drop('조식메뉴',axis=1,inplace=True)
test_data.drop('조식메뉴',axis=1,inplace=True)
```

```python
train_night=[]
for i in train_data['석식메뉴']:
    i=re.sub('\([^)]+\)','',i).replace('  ',' ')
    train_night.append(i)
```

```python
test_night=[]
for i in test_data['석식메뉴']:
    i=re.sub('\([^)]+\)','',i).replace('  ',' ')
    test_night.append(i)
```

```python
train_night_=[]
for i in range(len(train_data)):
    li_=[]
    li=train_night[i].split( )
    for j in li:
        if '쌀밥' in j:
            li_.append('밥')
        elif '현미밥' in j:
            li_.append('밥')
        elif '흑미밥' in j:
            li_.append('밥')
        elif '수수밥' in j:
            li_.append('밥')
        elif '귀리밥' in j:
            li_.append('밥')
        elif '찰현미밥' in j:
            li_.append('밥')
        elif '오곡밥' in j:
            li_.append('밥')
        elif '오곡밥' in j:
            li_.append('밥')
        else:
            li_.append(j)
    train_night_.append(li_)
```

```python
test_night_=[]
for i in range(len(test_data)):
    li_=[]
    li=test_night[i].split( )
    for j in li:
        if '쌀밥' in j:
            li_.append('밥')
        elif '현미밥' in j:
            li_.append('밥')
        elif '흑미밥' in j:
            li_.append('밥')
        elif '수수밥' in j:
            li_.append('밥')
        elif '귀리밥' in j:
            li_.append('밥')
        elif '찰현미밥' in j:
            li_.append('밥')
        elif '오곡밥' in j:
            li_.append('밥')
        elif '오곡밥' in j:
            li_.append('밥')
        else:
            li_.append(j)
    test_night_.append(li_)
```

```python
train_d=pd.DataFrame(train_night_)
test_d=pd.DataFrame(test_night_)
```

```python
train_d=train_d.iloc[:,:3]
test_d=test_d.iloc[:,:3]
```

```python
train_d = train_d.rename(columns={0: '밥', 1: '국',2:'메인'})
test_d = test_d.rename(columns={0: '밥', 1: '국',2:'메인'})
```

```python
cat_features = ['밥','국','메인']
for i in enumerate (cat_features):
    ca = i[1]
    train_d[ca] = train_d[ca].astype('category') 
```

```python
cat_features = ['밥','국','메인']
for i in enumerate (cat_features):
    ca = i[1]
    test_d[ca] = test_d[ca].astype('category') 
```

```python
train=pd.concat([train_data.iloc[:,:-2],train_data.iloc[:,-1]],axis=1)
test=test_data.iloc[:,:-2]
```

```python
train=train.drop(['석식메뉴'],axis=1)
```
