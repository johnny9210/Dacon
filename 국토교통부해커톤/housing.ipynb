```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
data = pd.read_table('./data/report.txt',thousands=',')
```

```python
data=data.iloc[:,:5]
```

```python
data
```

```python
v=data['자치구'].iloc[1:26]
for i in range(4):
    plt.figure(figsize=(20,20))
    plt.bar([j for j in range(25)], data['전체세대'].iloc[(26*i)+1:26*(i+1)])
    plt.xticks([j for j in range(25)],v,fontsize=10)
    plt.show()
```
