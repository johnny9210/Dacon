```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import re
```

```python
train_data=pd.read_csv('./data/train.csv')
```

```python
train_data.drop('조식메뉴',axis=1,inplace=True)
train_data
```

```python
afternoon=[]
night=[]
for i in train_data['중식메뉴']:
    i=re.sub('\([^)]+\)','',i).replace('  ',' ').replace('/',' ')
    afternoon.append(i)
    
for i in train_data['석식메뉴']:
    i=re.sub('\([^)]+\)','',i).replace('  ',' ').replace('/',' ')
    night.append(i)
```

```python
data=pd.DataFrame({'중식메뉴_':afternoon,'석식메뉴_':night})
```

```python
train_data=pd.concat([train_data,data],axis=1)
```

```python
train_data.drop(['중식메뉴','석식메뉴'],axis=1,inplace=True)
train_data=train_data[['일자','요일','본사정원수','본사휴가자수','본사출장자수','본사시간외근무명령서승인건수','현본사소속재택근무자수','중식메뉴_','석식메뉴_','중식계','석식계']]
train_data.head()
```
