```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import re
```

```python
train_data=pd.read_csv('./data/train.csv')
test_data=pd.read_csv('./data/test.csv')
```

```python
train_data.drop('조식메뉴',axis=1,inplace=True)
test_data.drop('조식메뉴',axis=1,inplace=True)
```

```python
train_lunch=[]
train_night=[]
for i in train_data['중식메뉴']:
    i=re.sub('\([^)]+\)','',i).replace('  ',' ')
    train_lunch.append(i)
    
for i in train_data['석식메뉴']:
    i=re.sub('\([^)]+\)','',i).replace('  ',' ')
    train_night.append(i)
```

```python
test_lunch=[]
test_night=[]
for i in test_data['중식메뉴']:
    i=re.sub('\([^)]+\)','',i).replace('  ',' ')
    test_lunch.append(i)
    
for i in test_data['석식메뉴']:
    i=re.sub('\([^)]+\)','',i).replace('  ',' ')
    test_night.append(i)
```

```python
train_lunch_=[]
for i in range(len(train_data)):
    li_=[]
    li=train_lunch[i].split( )
    for j in li:
        if '쌀밥' in j:
            li_.append('밥')
        elif '현미밥' in j:
            li_.append('밥')
        elif '흑미밥' in j:
            li_.append('밥')
        elif '수수밥' in j:
            li_.append('밥')
        elif '귀리밥' in j:
            li_.append('밥')
        elif '찰현미밥' in j:
            li_.append('밥')
        elif '오곡밥' in j:
            li_.append('밥')
        elif '오곡밥' in j:
            li_.append('밥')
        else:
            li_.append(j)
    train_lunch_.append(li_)
```

```python
test_night_=[]
for i in range(len(test_data)):
    li_=[]
    li=test_night[i].split( )
    for j in li:
        if '쌀밥' in j:
            li_.append('밥')
        elif '현미밥' in j:
            li_.append('밥')
        elif '흑미밥' in j:
            li_.append('밥')
        elif '수수밥' in j:
            li_.append('밥')
        elif '귀리밥' in j:
            li_.append('밥')
        elif '찰현미밥' in j:
            li_.append('밥')
        elif '오곡밥' in j:
            li_.append('밥')
        elif '오곡밥' in j:
            li_.append('밥')
        else:
            li_.append(j)
    test_night_.append(li_)
```

```python
train_d=pd.DataFrame(train_lunch_)
test_d=pd.DataFrame(test_night_)
```

```python
train_d=train_d.iloc[:,:3]
test_d=test_d.iloc[:,:3]
```

```python
train_d = train_d.rename(columns={0: '밥', 1: '국',2:'메인'})
test_d = test_d.rename(columns={0: '밥', 1: '국',2:'메인'})
```

```python
cat_features = ['밥','국','메인']
for i in enumerate (cat_features):
    ca = i[1]
    train_d[ca] = train_d[ca].astype('category') 
```

```python
cat_features = ['밥','국','메인']
for i in enumerate (cat_features):
    ca = i[1]
    test_d[ca] = test_d[ca].astype('category') 
```

```python
train=train_data.iloc[:,:-1]
test=test_data.iloc[:,:-1]
```

```python
train=train.drop(['본사시간외근무명령서승인건수','석식메뉴'],axis=1)
test=test.drop(['본사시간외근무명령서승인건수'],axis=1)
```

```python
train=pd.concat([train,train_d],axis=1)
test=pd.concat([test,test_d],axis=1)
```

```python
train=train.drop(['중식메뉴'],axis=1)
test=test.drop(['중식메뉴'],axis=1)
```

```python
train=train[['일자','요일','본사정원수','본사휴가자수','본사출장자수','현본사소속재택근무자수','밥','국','메인','중식계']]
```
